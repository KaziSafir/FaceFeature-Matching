```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read image
Julia = cv2.imread('Julia-Roberts.jpg',0)
Nelson = cv2.imread('NelsonMandela.jpg',0)
world = cv2.imread('world_conference.webp',0)
```

```python
#transform image into black and white
plt.imshow(Julia,cmap='gray')
```

```python
#transform image into black and white
plt.imshow(Nelson,cmap='gray')
```

```python
#transform image into black and white
plt.imshow(world,cmap='gray')
```

```python
#importing haarcascade_frontalface_default.xml which is face detection xml
facedetect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#creating a class called face that will implament the face detection eml
def face(img):
  faceimg = img.copy()
  facerects = facedetect.detectMultiScale(faceimg)
  for (x,y,w,h) in facerects:
    cv2.rectangle(faceimg, (x,y),(x+w,y+h), (255,255,255), 10)
  return faceimg
```

```python
#Face detection for julia
result = face(Julia)
plt.imshow(result,cmap='gray')
```

```python
#Face detection for Nelson
result = face(Nelson)
plt.imshow(result,cmap='gray')
```

```python
#since world conference is made up of multiple people and faces it will need a new face detecting code
#this code will detect multiple faces
def multiface(img):
  faceimg = img.copy()
  facerects = facedetect.detectMultiScale(faceimg,scaleFactor=1.2, minNeighbors=2)
  for (x,y,w,h) in facerects:
    cv2.rectangle(faceimg, (x,y),(x+w,y+h), (255,255,255), 2)
  return faceimg

#detecting multi face in world conference
result = multiface(world)
plt.imshow(result,cmap='gray')
#some people will not be detected because they're not facing forward
```

```python
#to detect eyes we need to use the haarcascade_eye.xml
#imforting the file
eyedetect = cv2.CascadeClassifier("haarcascade_eye.xml")
def eyes(img):
  faceimg = img.copy()
  eyes = eyedetect.detectMultiScale(faceimg)
  for (x,y,w,h) in eyes:
    cv2.rectangle(faceimg, (x,y),(x+w,y+h), (255,255,255), 10)
  return faceimg
```

```python
#detecting eyes for julia
result = eyes(Julia)
plt.imshow(result,cmap='gray')
```

```python
#detecting eyes for Nelson
result = eyes(Nelson)
plt.imshow(result,cmap='gray')
#no eyes will be detected because he is squinting
```

```python
#detecting eyes for world conference
result = eyes(world)
plt.imshow(result,cmap='gray')
#no eyes will detect because the qulity of the photo
```
